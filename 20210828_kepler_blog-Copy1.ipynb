```python
import pandas as pd
import geopandas as gpd

# Shape file path
path = 'data/tambon/tha_admbnda_adm3_rtsd_20190221.shp'

# read shapefile into geo pandas data frame
df_map = gpd.read_file(path)

# check data
df_map.head()
```

```python
df_map[['ADM3_EN', 'ADM2_EN', 'ADM1_EN', 'geometry']]
```

```python
# sales data
df_sale_by_subdistrict = pd.read_csv('data/sales_by_province.csv')
df_sale_by_subdistrict
```

```python
# merge sales data to shapefile data
dfkepler = df_sale_by_subdistrict.merge(df_map, left_on = ['province', 'district', 'subdistrict'], right_on = ['ADM1_EN','ADM2_EN','ADM3_EN'])
dfkepler

# export to plot in Kepler
dfkepler.to_csv('output/th_sales_map.csv', index = False)
```
